```python
heart_disease=pd.read_csv("heart-disease.csv")
heart_disease
```

```python
heart_disease.plot.hist(figsize=(10,30),subplots=True)
```

```python
over_50=heart_disease[heart_disease["age"] >50]
over_50
```

```python
len(over_50)
```

```python
#pyplot method
over_50.plot(kind='scatter',
            x='age',
            y="chol",
            c="target")
```

```python
#oo method
fix, ax= plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x="age",
            y="chol",
            c="target",
            ax=ax)
```

```python
#OO method
fig,ax= plt.subplots(figsize=(10,6))
# plot the data
scatter=ax.scatter(x=over_50["age"],
          y=over_50["chol"],
          c=over_50["target"])
#customize the plot 
ax.set(title="distrbution chol with age",
      xlabel="Age",
      ylabel="cholestrol")
# add a legend 
ax.legend(*scatter.legend_elements(),title="Target")

#add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--');
```

```python
over_50
```

```python
#Set the style
plt.style.use('seaborn-whitegrid')

# subplot of chol, age, thalach
fig,(ax0,ax1)= plt.subplots(nrows=2,
                           ncols=1,
                           figsize=(8,10),
                           sharex=True)
# Add data to ax0
scatter =ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap="winter")

#Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
       ylabel="cholesterol")

# change the x axis limet
ax0.set_xlim([50,80])
ax0,ax1.set_ylim([60,200])
#add legend
ax0.legend(*scatter.legend_elements(),title="Target")

#add a horizontal line
ax0.axhline(over_50["chol"].mean(),
            linestyle='--')

#add data to ax1
scatter =ax1.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"],
                    cmap="winter")
#Customize ax
ax1.set(title="Heart Disease and thalach Levels",
       xlabel="Age",
       ylabel="thalach")

#add legend
ax1.legend(*scatter.legend_elements(),title="Target")

#add a horizontal line
ax1.axhline(over_50["thalach"].mean(),
            linestyle='--')
# add a title to the figure
fig.suptitle("Heart Diseas Analysis", fontsize=16 ,fontweight="bold")
```
